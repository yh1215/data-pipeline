```python
import pandas as pd

data = {
    'student': ['Allen', 'Ted', 'Jack'],
    'exam_1': [32, 31, 33],
    'exam_2': [76, 78, 77],
    'exam_3': [88, 91, 92]
}

df = pd.DataFrame(data)

df
```

```python
melted_df = pd.melt(df,
                    id_vars= ['student'],
                    var_name='exam',
                    value_name='score')

melted_df
```

```python
data = {
    'classroom': ['A', 'A', 'B', 'B'],
    'student': ['Allen', 'Ted', 'Jack', 'Mary'],
    'exam_1': [32, 31, 33, 35],
    'exam_2': [76, 78, 77, 75],
    'exam_3': [88, 91, 92, 85]
}

df = pd.DataFrame(data)

df
```

```python
# melt with multipul cols
melted_df =pd.melt(df,
                   id_vars=['classroom', 'student'],
                   var_name='exam',
                   value_name= 'score'
                    )

melted_df
```
