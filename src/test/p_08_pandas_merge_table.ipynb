```python
import pandas as pd
```

```python
file_path_staff = '/workspaces/demo-devcontainer-main/resource/staff.csv'
file_path_staff_detail = '/workspaces/demo-devcontainer-main/resource/staff_detail.csv'

df_staff = pd.read_csv(file_path_staff)
df_staff_detail = pd.read_csv(file_path_staff_detail)
```

```python
df_staff
```

```python
df_staff_detail
```

```python
merge_df = df_staff.merge(df_staff_detail, how='inner', on='ID')
merge_df
```

```python
file_path_staff_detail_different_col= '/workspaces/demo-devcontainer-main/resource/staff_detail_different_col.csv'

df_staff_detail_different_col = pd.read_csv(file_path_staff_detail_different_col)

df_staff_detail_different_col
```

```python
# merge df with different key
merge_df = pd.merge(
    df_staff_detail_different_col,
    df_staff,
    how='left',
    left_on='EmpNum',
    right_on='ID'
)

merge_df
```

```python
file_path_staff_samecol = '/workspaces/demo-devcontainer-main/resource/staff_samecol.csv'
file_path_staff_detail = '/workspaces/demo-devcontainer-main/resource/staff_detail.csv'

staff_samecol = pd.read_csv(file_path_staff_samecol)
staff_detail = pd.read_csv(file_path_staff_detail)
```
