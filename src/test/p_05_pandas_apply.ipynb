```python
import pandas as pd
```

```python
file_path = '/workspaces/demo-devcontainer-main/src/pandas_demo/ptt.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
"""drop str part and keep int in columns: 推: 25 -> 25"""

def tmp_apply(input_str: str) -> int:
    """split"""
    return int(input_str.split(': ')[1])
    """replace and strip(drop space at head and tail)
    return int(input_str.replace('推: ', ''), strip())"""
```

```python
df['推'].apply(tmp_apply)
```

```python
df['推'] = df['推'].apply(tmp_apply)

df.head()
```

```python
df['噓'] = df['噓'].apply(
    lambda input_str: int(input_str.replace('噓: ', '').strip())
)

df.head()
```

```python
df['分數'] = df['分數'].apply(
    lambda input_str: int(input_str.replace('分數: ', '').strip())
)

df.head()
```

```python
df["時間"] = df["時間"].apply(
    lambda input_str: input_str.replace("時間: ", "").strip()
)

df.head()
```

```python
from datetime import datetime
from datetime import date
```

```python
df['time'] = pd.to_datetime(df['時間'])
```
